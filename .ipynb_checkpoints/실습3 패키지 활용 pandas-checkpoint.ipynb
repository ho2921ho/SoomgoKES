```python
import pandas as pd
```

```python
df = pd.read_csv('data/auction.csv')
```

```python
result = pd.read_csv('data/auction_result.csv')
```

```python
df
```

### 1. 건물평수 칼럼의 자료형을 int로 변환

```python
# 결과 체크
sum(df['건물평수'] == result['건물평수']) / len(df['건물평수'])
```

### 2. 조회수  칼럼의 자료형을 int로 변환

```python
# 결과 체크
sum(df['조회수'] == result['조회수']) / len(df['조회수'])
```

### 3. 감정가와 최저가 칼럼의 자료형을 int로 변환

```python
# 결과 체크
sum(df['감정가'] == result['감정가']) / len(df['감정가'])
```

```python
# 결과 체크
sum(df['최저가'] == result['최저가']) / len(df['최저가'])
```

### 4. 해당 물건의 할인율을 계산하고 '할인율' 칼럼을 추가(소수점 셋째자리에서 반올림)

```python
# 결과 체크
sum(df['할인율'] == result['할인율']) / len(df['할인율'])
```
